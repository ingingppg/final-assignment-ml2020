# Modeling

> Model Function

```python
def prepare_data(location):
    
    #select location
    locations = covid.loc[covid_scaled['location'] == location]

    #split train data
    train_data = locations.loc[locations['date'] < '2021-04-12']
    train_data_y = train_data[['total_cases']]
    train_data_X = train_data.drop(['total_cases','location','date', 'tests_units', 'iso_code', 'continent'], axis=1)

    #spilt test data
    test_data = locations.loc[(locations['date'] >= '2021-04-12') & (locations['date'] < '2021-04-19')]
    test_data_y = test_data[['total_cases']]
    test_data_X = test_data.drop(['total_cases','location','date', 'tests_units', 'iso_code', 'continent'], axis=1)

    return train_data_X, train_data_y, test_data_X, test_data_y
```

```python
from sklearn.model_selection import train_test_split

def splitdata(train_data_X, train_data_y):

    X_train, X_test, y_train, y_test = train_test_split(train_data_X, train_data_y, test_size=0.3, random_state=42)

    return X_train, X_test, y_train, y_test
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.pipeline import make_pipeline

def multiregression(X_train, X_test, y_train, y_test, test_data_X):

    #call model
    model = make_pipeline(StandardScaler(),
                          LinearRegression())

    #fit model with train
    model.fit(X_train, y_train)

    # Predict with test
    y_pred = model.predict(X_test)

    #MSE
    mse = mean_squared_error(y_test, y_pred)

    #R2
    r2 = r2_score(y_test, y_pred)

    #Predict with real data
    y_pred = model.predict(test_data_X)

    return mse , r2, y_pred
```

```python
from sklearn import ensemble

def GDB(X_train, X_test, y_train, y_test, test_data_X):

    #call model
    model = make_pipeline(StandardScaler(),
                          ensemble.GradientBoostingRegressor())

    #fit model with train
    model.fit(X_train, y_train.values.ravel())

    # Predict with test
    y_pred = model.predict(X_test)

    #MSE
    mse = mean_squared_error(y_test, y_pred)

    #R2
    r2 = r2_score(y_test, y_pred)

    #Predict with real data
    y_pred = model.predict(test_data_X)

    return mse , r2, y_pred
```

```python
from sklearn.linear_model import SGDRegressor

def SGD(X_train, X_test, y_train, y_test, test_data_X):

    #call model
    model = make_pipeline(StandardScaler(),
                          SGDRegressor(max_iter=1000))

    #fit model with train
    model.fit(X_train, y_train.values.ravel())

    # Predict with test
    y_pred = model.predict(X_test)

    #MSE
    mse = mean_squared_error(y_test, y_pred)

    #R2
    r2 = r2_score(y_test, y_pred)

    #Predict with real data
    y_pred = model.predict(test_data_X)

    return mse , r2, y_pred
```

## World

```python
# World

train_data_X, train_data_y, test_data_X, test_data_y = prepare_data('World')

X_train, X_test, y_train, y_test = splitdata(train_data_X, train_data_y)

mlg_mse, mlg_r2, mlg_y_pred = multiregression(X_train, X_test, y_train, y_test, test_data_X)

gdb_mse, gdb_r2, gdb_y_pred = GDB(X_train, X_test, y_train, y_test, test_data_X)

sgd_mse, sgd_r2, sgd_y_pred = SGD(X_train, X_test, y_train, y_test, test_data_X)
```

```python
print("--- Multi Regression ---")
print("MSE on train set: {:.5f}".format(mlg_mse))
print("R2 score : {:.2f}%".format(mlg_r2*100))

print("\n--- Gradient Boosting ---")
print("MSE on train set: {:.5f}".format(gdb_mse))
print("R2 score : {:.2f}%".format(gdb_r2*100))

print("\n--- Stochastic Gradient Descent ---")
print("MSE on train set: {:.5f}".format(sgd_mse))
print("R2 score : {:.2f}%".format(sgd_r2*100))
```
